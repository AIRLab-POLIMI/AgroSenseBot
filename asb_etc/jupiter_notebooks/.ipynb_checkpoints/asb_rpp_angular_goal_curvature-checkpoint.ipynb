```python
from numpy import *
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [7, 7]
from matplotlib.patches import Arc

# x_g, y_g, t: goal pose (t is theta)
# x_p, y_p: intersection point between goal pose y-axis and robot pose y-axis
# x_a, y_a: angle goal pose in ASB RPP, new pose translated on the goal pose's y-axis such that 
#     the final angle on the lookahead arc is the same as the goal angle
# c: curvature that leads the robot to the x_a, y_a pose

def f(x_g, y_g, t):
    x_p = 0
    y_p = (y_g + x_g / tan(t)) if abs(t) > 0 else inf
    c = tan(t) / (x_g + y_g * tan(t))
    y_p = 1 / c if abs(t) > 0 else inf
    x_a = y_g * sin(t) + x_g * cos(t)
    y_a = x_g * cos(t) * tan(t/2) + y_g * (1 - cos(t))
    
    print(f"x_g = {x_g:.4f} y_g = {y_g:.4f}, t = {t/pi:.4f} pi")
    print(f"x_p = {x_p:.4f} y_p = {y_p:.4f}")
    print(f"x_a = {x_a:.4f} y_a = {y_a:.4f}")

    return x_g, y_g, t, x_p, y_p, x_a, y_a

def plot(x_g, y_g, t, x_p, y_p, x_a, y_a):
    fig, ax = plt.subplots()
    
    plt.plot([x_g, x_a], [y_g, y_a], color='lime')
    plt.scatter(x_p, y_p, color='lime')
    
    plt.quiver(x_g, y_g, cos(t), sin(t), color='slateblue')
    
    plt.quiver(x_a, y_a, cos(t), sin(t), color='red')

    if abs(y_p) < inf:
        r = abs(y_p)
        t_1 = pi/2 if y_p < 0 else -pi/2
        t_2 = t_1 + t
        arc_angles = linspace(min(t_1, t_2), max(t_1, t_2), 100)
        arc_xs = x_p + r * cos(arc_angles)
        arc_ys = y_p + r * sin(arc_angles)
        plt.plot(arc_xs, arc_ys, color='red', lw = 3)
    else:
        plt.plot([0, x_a], [0, y_a], color='red', lw = 3)

    plt.quiver(0, 0, 1, 0, color='black')
    
    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    plt.grid()
    # ax.axhline(y=0, color='k')
    # ax.axvline(x=0, color='k')
    ax.set_aspect('equal', adjustable='box')
    # plt.figure().set_figwidth(15)
    plt.show()

plot(*f(x_g=0.75, y_g=0.0, t=0))

plot(*f(x_g=0.75, y_g=0.0, t=-0.2 * pi))
plot(*f(x_g=0.75, y_g=0.0, t=0.2 * pi))

plot(*f(x_g=0.75, y_g=0.5, t=-0.01*pi))
plot(*f(x_g=0.75, y_g=0.5, t=0))
plot(*f(x_g=0.75, y_g=0.5, t=0.01*pi))

plot(*f(x_g=0.75, y_g=-0.5, t=-0.01*pi))
plot(*f(x_g=0.75, y_g=-0.5, t=0))
plot(*f(x_g=0.75, y_g=-0.5, t=0.01*pi))

plot(*f(x_g=-0.75, y_g=0.5, t=-0.01*pi))
plot(*f(x_g=-0.75, y_g=0.5, t=0))
plot(*f(x_g=-0.75, y_g=0.5, t=0.01*pi))

plot(*f(x_g=-0.75, y_g=-0.5, t=-0.01*pi))
plot(*f(x_g=-0.75, y_g=-0.5, t=0))
plot(*f(x_g=-0.75, y_g=-0.5, t=0.01*pi))

plot(*f(x_g=0.75, y_g=0.5, t=pi/2 - 0.01*pi))
plot(*f(x_g=0.75, y_g=0.5, t=pi/2 - 0.1*pi))

plot(*f(x_g=0.75, y_g=0.5, t=pi/4 + 0.1*pi))
plot(*f(x_g=0.75, y_g=0.5, t=pi/4))
plot(*f(x_g=0.75, y_g=0.5, t=pi/4 - 0.1*pi))

plot(*f(x_g=0.75, y_g=0.5, t=-0.5 * pi))
plot(*f(x_g=0.75, y_g=0.5, t=-0.4 * pi))
plot(*f(x_g=0.75, y_g=0.5, t=-0.3 * pi))
plot(*f(x_g=0.75, y_g=0.5, t=-0.25 * pi))
plot(*f(x_g=0.75, y_g=0.5, t=-0.2 * pi))

plot(*f(x_g=0.75, y_g=0.5, t=-30 * pi/180))
plot(*f(x_g=-0.75, y_g=0.5, t=-30 * pi/180))

plot(*f(x_g=0.75, y_g=-0.5, t=30 * pi/180))
plot(*f(x_g=0.75, y_g=-0.5, t=10 * pi/180))
plot(*f(x_g=0.75, y_g=-0.5, t=-10 * pi/180))
```
